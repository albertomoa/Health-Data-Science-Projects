## 2.2 Importing Libraries

```python
# import necessary libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

## 2.3 Search for Target Protein
### 2.3.1 Target search for Coronavirus

```python
# target search for coronavirus
target = new_client.target
target_query = target.search('coronavirus')
targets = pd.DataFrame.from_dict(target_query)
targets
```

### 2.3.2 Select and Retrieve Data for SARS Coronavirus 3C-like Proteinase (sevent entry)

Seventh entry (which corresponds to the target protein, coronavirus 3C-like proteinase) to the **selected_target** variable

```python
# save target chembl id into selected_target variable
selected_target = targets.target_chembl_id[6]
selected_target
```

Here, we will retrieve only bioactivity data for *coronavirus 3C-like proteinase* (CHEMBL3972) that are reported as $IC_{50}$ values in nM(nanomolar) unit.

```python
# retrieve bioactivity data and save it to res variable
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
# save res data into dataframe
df = pd.DataFrame.from_dict(res)
df
```

```python
# check data variant on standard_type column
df.standard_type.unique()
```

Finally, resulting bioactivity data into CSV file **bioactivity_data.csv**

```python
df.to_csv('bioactivity_data.csv', index=False)
```

## 2.5 Data Pre-Processing of the Bioactivity Data

### 4.1 Labelling Compounds as either being Active, Inactive, or Intermediate

The bioactivity data is in the $IC_{50}$ unit. Compounds having standard values of less than 1,000 nM will be considered to be **active** while those greater than 10,000 nM will be **inactive**. As for those values in between 1,000 and 10,000 nM will be referred to as **intermediate**.

```python
# labelling compouds based on standard_value
# save new label into bioactivity_class variable
bioactivity_class = []
for i in df.standard_value:
    if float(i) >= 10000:
        bioactivity_class.append('inactive')
    elif float(i) <= 1000:
        bioactivity_class.append('active')
    else:
        bioactivity_class.append('intermediate')
```

## 4.1 Retrive necessary information from `df` dataset

```python
# gather neccesary information from df dataset
selection = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
df2 = df[selection]
```

### Combine `Biocativity_class` lists above and df2 dataset

```python
df_temp = pd.concat([df2, pd.Series(bioactivity_class)], axis=1)
df_temp
```

```python
# rename the column
df3 = df_temp.rename(columns={0: 'bioactivity_data'})
df3
```

```python
df3.to_csv('bioactivity_preprocessed_data.csv', index=False)
```
